
# Primer punto Distribuciones discretas.
Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números:
Grafique los números aleatorios,
Ordene de menor a mayor y grafique,
Encuentre los valores teoricos de media y varianza, y comparelos con los estimados.

```python
#Para todas las simulaciones:
N=10000000
m=300
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# bernoulli con cierto p entre 0 y 1
p=0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
r = bernoulli.rvs(p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos por bernoulli con p="+str(p))
plt.show()
```

```python
r.sort()
plt.plot(r,'r--')
plt.title(str(N)+" puntos distribuidos por bernoulli ordenados con p="+str(p))
plt.grid()
plt.show()
```

```python
plt.hist(r, color = 'red',bins=m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos por bernoulli con p="+str(p))
plt.grid()
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución de bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la distribución bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```python
# binomial con cierto p entre 0 y 1 y cierto n
n,p = 10, 0.5
from scipy.stats import binom
mean, var, skew, kurt = binom.stats(n,p, moments='mvsk')
r = binom.rvs(n,p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos por binomnial con p="+str(p))
plt.show()
#***************************************************************
r.sort()
plt.plot(r,'r--')
plt.title(str(N)+" puntos distribuidos por binomial ordenados con p="+str(p))
plt.grid()
plt.show()
#***************************************************************
plt.hist(r, color='red',bins=m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos por binomial con p="+str(p))
plt.grid()
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución binomial con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la distribución binomial con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```python
# Normal de media 5 y desviación estandar 0.5
u,std=5,0.5
from scipy.stats import norm
mean, var, skew, kurt = norm.stats(moments='mvsk',loc=u,scale=std)
r = norm.rvs (loc=u,scale=std, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normalmente con media="+str(u)+" y std="+str(std))
plt.show()
#*******************************************************************************
r.sort()
plt.plot(r,'r--')
plt.title(str(N)+" puntos distribuidos normalmente ordenados con p="+str(p))
plt.grid()
plt.show()
#*******************************************************************************
plt.hist(r, bins = m, color = 'red')
plt.title("Histograma de los "+str(N)+" puntos distribuidos normalmente con u="+str(u)+ " y std= "+str(std))
plt.grid()
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```
